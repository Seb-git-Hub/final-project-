```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "paintings_master.csv"

# Read paintigns master File and store into Pandas data frame
paintings_data = pd.read_csv(file_to_load)
```

```python
Total_Artist=pd.DataFrame(paintings_data.Artist_id.value_counts())
```

```python
Total_Artist.rename(columns={'Artist_id': 'count'}, inplace=True)
Total_Artist.index.rename('artist_id', inplace=True)
Total_Artist
```

```python
df = pd.DataFrame(Total_Artist)
#df = df.sort_values(['A', 'C'], ascending=[1, 1])
df.sort_values(['artist_id','count'], ascending=[0,0])
#head()
print(df)
#df.head(5)
```
